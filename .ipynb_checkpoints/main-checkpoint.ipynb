# loan_term and Monthly mortgage payment 
Loan Term: Often 20-30 years, Initial Tilgung(the annual repayment rate): Around 2-3%

```python
def calculate_amortization(loan_amount, annual_interest_rate, monthly_payment=None, loan_term=None):
    """
    Calculate loan details depending on given parameters.
    
    Required Parameters:
    - loan_amount: The principal borrowed.
    - annual_interest_rate: Annual interest rate (e.g., 0.02 for 2% per year).

    Optional Parameters:
    Provide EITHER monthly_payment OR loan_term (in years), not both:
    - monthly_payment: Fixed monthly payment you will make.
    - loan_term: Desired loan term in years.

    Returns:
    A dictionary containing:
    - monthly_payment
    - total_interest_paid
    - total_months (term of the loan in months)
    - initial_tilgung_rate (annual percentage of original principal)
    """

    monthly_interest_rate = annual_interest_rate / 12.0

    if monthly_payment is None and loan_term is not None:
        # Scenario B: Known loan_term, need to find monthly_payment
        total_months = loan_term * 12
        # Monthly payment formula for annuity: P = r * PV / (1 - (1+r)^(-n))
        # where:
        # P = monthly_payment
        # r = monthly_interest_rate
        # PV = loan_amount
        # n = total_months
        if monthly_interest_rate > 0:
            monthly_payment = (monthly_interest_rate * loan_amount) / (1 - (1 + monthly_interest_rate)**(-total_months))
        else:
            # If interest is zero, payment is just principal/term
            monthly_payment = loan_amount / total_months

    elif monthly_payment is not None and loan_term is None:
        # Scenario A: Known monthly_payment, we have to derive the loan term by simulation
        # We'll iterate month by month until the loan is paid off.
        pass
    else:
        raise ValueError("You must provide exactly one of (monthly_payment or loan_term).")

    # If monthly_payment was given and loan_term was None, we need to find the loan term by simulation
    if loan_term is None:
        remaining_balance = loan_amount
        months = 0
        while remaining_balance > 0:
            months += 1
            interest = remaining_balance * monthly_interest_rate
            principal = monthly_payment - interest
            if principal <= 0:
                raise ValueError("Monthly payment is too low to cover even the interest. Increase the monthly payment.")
            remaining_balance -= principal
            if months > 1000*12:
                # Safety check: if it takes more than 1000 years, break to avoid infinite loops
                break
        total_months = months

    else:
        # If loan_term was given, we have total_months already
        total_months = loan_term * 12

    # Once we have monthly_payment and total_months, we can calculate total interest and initial tilgung rate
    # We re-simulate the payment schedule
    remaining_balance = loan_amount
    total_interest_paid = 0.0
    for m in range(1, total_months + 1):
        interest = remaining_balance * monthly_interest_rate
        principal = monthly_payment - interest
        if principal < 0:
            raise ValueError("Monthly payment not sufficient to cover interest.")
        total_interest_paid += interest
        remaining_balance -= principal
        if remaining_balance < 0:
            remaining_balance = 0

    # Initial tilgung rate calculation:
    # The initial tilgung is how much principal we pay in the first month as a % of the original loan amount, annualized.
    # principal_first_month = monthly_payment - (loan_amount * monthly_interest_rate)
    # initial_tilgung_rate = (principal_first_month * 12) / loan_amount
    principal_first_month = monthly_payment - (loan_amount * monthly_interest_rate)
    initial_tilgung_rate = (principal_first_month * 12) / loan_amount

    return {
        "monthly_payment": monthly_payment,
        "total_interest_paid": total_interest_paid,
        "total_months": total_months,
        "initial_tilgung_rate": initial_tilgung_rate
    }

```

```python
# Example Usage:
# SCENARIO A:
# Known monthly payment scenario
print("Scenario A: Known Monthly Payment")
monthly_payment_given = 2300.0
result_a = calculate_amortization(
    loan_amount=loan_amount,
    annual_interest_rate=loan_interest_rate,
    monthly_payment=monthly_payment_given,
    loan_term=None
)
print(f"Monthly Payment: €{result_a['monthly_payment']:.2f}")
print(f"Total Interest Paid: €{result_a['total_interest_paid']:.2f}")
print(f"Loan Term: {result_a['total_months']} months (~{result_a['total_months']/12:.1f} years)")
print(f"Initial Tilgung Rate: {result_a['initial_tilgung_rate']*100:.2f}%\n")
```

```python
# SCENARIO B:
# Known loan term scenario
print("Scenario B: Known Loan Term")
loan_term_given = 30  # 30 years
result_b = calculate_amortization(
    loan_amount=loan_amount,
    annual_interest_rate=loan_interest_rate,
    monthly_payment=None,
    loan_term=loan_term_given
)
print(f"Monthly Payment: €{result_b['monthly_payment']:.2f}")
print(f"Total Interest Paid: €{result_b['total_interest_paid']:.2f}")
print(f"Loan Term: {result_b['total_months']} months")
print(f"Initial Tilgung Rate: {result_b['initial_tilgung_rate']*100:.2f}%")
```

# Model 1 buying versus renting
Down payment are all invested on e.g ETFs and has a investment_return_rate e.g 4 ~ 10 %

```python
loan_term = 25
```

```python
years = loan_term
```

```python
# Monthly mortgage payment calculation
monthly_payment = npf.pmt(monthly_loan_interest_rate, loan_term * 12, -loan_amount)
monthly_payment
```

```python
# Initialize lists to store annual values
buy_net_worth = []
rent_net_worth = []
months = years * 12
months_list = []
```

```python
for month in range(months):
    # Current property value each month
    monthly_property_value = purchase_price * ((1 + property_appreciation_rate/12)**month)

    # Monthly costs
    monthly_maintenance = (purchase_price * maintenance_rate)/12
    monthly_taxes = property_taxes / 12
    monthly_buying_cost = monthly_payment + monthly_maintenance + monthly_taxes

    # Monthly rent
    monthly_rent = initial_rent * ((1 + rent_inflation_rate/12) ** month)

    # Calculate monthly savings for renter scenario
    monthly_savings = monthly_buying_cost - monthly_rent
    
    if monthly_savings < 0:
        print(f"At month {month}, renting is no longer cheaper than buying. Stopping simulation.")
        break
        
    # Investment returns from the saved down payment and annual difference
    if month == 0:
        invested_capital = down_payment
    else:
        # Apply monthly investment returns
        invested_capital += monthly_savings

        monthly_investment_return = invested_capital * (investment_return_rate / 12)
        invested_capital += monthly_investment_return

    # Update cumulative cash flow for renting
    rent_net_worth.append(invested_capital)

    # Calculate equity for buying scenario
    # Outstanding balance after month payments:
    # npf.fv(rate, nper, pmt, pv) gives future value; for loan, pv is loan_amount neg.
    outstanding_balance = npf.fv(monthly_loan_interest_rate, month+1, monthly_payment, -loan_amount)
    equity = monthly_property_value - outstanding_balance
    buy_cumulative_cash_flow = equity - inital_payment
    buy_net_worth.append(buy_cumulative_cash_flow)
    
    # Store year
    months_list.append(month)
```

```python
# Convert months to years
years_list = [month / 12 for month in months_list]

# Create DataFrame for visualization
data = pd.DataFrame({
    'Years': years_list,
    'Buying Net Worth (€)': buy_net_worth,
    'Renting Net Worth (€)': rent_net_worth
})

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(data['Years'], data['Buying Net Worth (€)'], label='Buying')
plt.plot(data['Years'], data['Renting Net Worth (€)'], label='Renting')
plt.title('Net Worth Over Time: Buying vs Renting')
plt.xlabel('Years')
plt.ylabel('Net Worth (€)')
plt.legend()
plt.grid(True)
plt.show()
```

# Model 2 buying versus renting
Only half of the down payment are invested on e.g ETFs and has a investment_return_rate e.g 4 ~ 10 %. The other half are invested on bank savings and has a return of loan_interest_rate - 2 % 

```python
# Initialize lists to store annual values
buy_net_worth = []
rent_net_worth = []
months = years * 12
months_list = []
```

```python
for month in range(months):
    # Current property value each month
    monthly_property_value = purchase_price * ((1 + property_appreciation_rate/12)**month)

    # Monthly costs
    monthly_maintenance = (purchase_price * maintenance_rate)/12
    monthly_taxes = property_taxes / 12
    monthly_buying_cost = monthly_payment + monthly_maintenance + monthly_taxes

    # Monthly rent
    monthly_rent = initial_rent * ((1 + rent_inflation_rate/12) ** month)

    # Calculate monthly savings for renter scenario
    monthly_savings = monthly_buying_cost - monthly_rent
    
    if monthly_savings < 0:
        print(f"At month {month}, renting is no longer cheaper than buying. Stopping simulation.")
        break
        
    # Investment returns from the saved down payment and annual difference
    if month == 0:
        invested_capital_total = down_payment
        invested_capital_ETF = invested_capital_total/2
        invested_capital_savings = invested_capital_total /2

    else:
        # Apply monthly investment returns
        invested_capital_ETF += monthly_savings
        monthly_investment_return1 = invested_capital_ETF * (investment_return_rate / 12) 
        monthly_investment_return2 = invested_capital_savings * ((loan_interest_rate-0.02) / 12) 
        invested_capital_ETF += monthly_investment_return1
        invested_capital_savings += monthly_investment_return2
        invested_capital_total = invested_capital_ETF + invested_capital_savings
    # Update cumulative cash flow for renting
    rent_net_worth.append(invested_capital_total)

    # Calculate equity for buying scenario
    # Outstanding balance after month payments:
    # npf.fv(rate, nper, pmt, pv) gives future value; for loan, pv is loan_amount neg.
    outstanding_balance = npf.fv(monthly_loan_interest_rate, month+1, monthly_payment, -loan_amount)
    equity = monthly_property_value - outstanding_balance
    buy_cumulative_cash_flow = equity - inital_payment
    buy_net_worth.append(buy_cumulative_cash_flow)
    
    # Store year
    months_list.append(month)
```

```python
# Convert months to years
years_list = [month / 12 for month in months_list]

# Create DataFrame for visualization
data = pd.DataFrame({
    'Years': years_list,
    'Buying Net Worth (€)': buy_net_worth,
    'Renting Net Worth (€)': rent_net_worth
})

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(data['Years'], data['Buying Net Worth (€)'], label='Buying')
plt.plot(data['Years'], data['Renting Net Worth (€)'], label='Renting')
plt.title('Net Worth Over Time: Buying vs Renting')
plt.xlabel('Years')
plt.ylabel('Net Worth (€)')
plt.legend()
plt.grid(True)
plt.show()
```
